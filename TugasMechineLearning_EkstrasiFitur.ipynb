Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,
1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
2. Lakukan proses encoding pada kolom "diagnosis".
3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.
4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd

# Replace 'your_file.csv' with the name of your CSV file
df = pd.read_csv('/content/drives/MyDrive/Dataset/wbc.csv')

# Now you can work with the DataFrame 'df' containing your data
df.head()
```

### 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
columns_to_exclude = ['Unnamed: 32', 'id']
data_filtered = df.drop(columns=columns_to_exclude)
```

```python
data_filtered.head()
```

### 2. Lakukan proses encoding pada kolom "diagnosis".

```python
label_encoder = LabelEncoder()
data_filtered['diagnosis'] = label_encoder.fit_transform(data_filtered['diagnosis'])
```

```python
df.head()
```

### 3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
scaler = StandardScaler()
numeric_columns = df.columns.difference(['diagnosis'])
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
df.head()
```

### 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
diagnosis_only = df['diagnosis']
data_train, data_test = train_test_split(df, test_size=0.2, random_state=0, stratify=diagnosis_only)

# Cek masing-masing ukuran data

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{data_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{data_test.diagnosis.value_counts()}')
```
