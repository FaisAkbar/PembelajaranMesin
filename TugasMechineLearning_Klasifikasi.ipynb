```python
# Menggunakan data spam.csv
# conect to google drive
from google.colab import drive
drive.mount('/content/drive')

# spesifiksi encoding diperlukan karena data tidak menggunakan UTF-8
dt = pd.read_csv('/content/drive/MyDrive/Dataset/spam.csv', encoding='latin-1')

dt = dt.drop(dt.iloc[:,2:], axis=1)
dt.head()
```

```python
# Fitur CountVectorizer dengan mengaktifkan stop_words
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(dt['v2'], dt['v1'], test_size=0.2, random_state=42)
# Create the CountVectorizer object
vectorizer = CountVectorizer(stop_words='english')
```

```python
# Transform the training and testing data
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
# Create the Multinomial Naive Bayes model
model = MultinomialNB()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print(classification_report(y_test, y_pred))
```

```python
# Menggunakan data spam.csv
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report

dt.head()
```

```python
# Fitur TF-IDF dengan mengaktifkan stop_words
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(dt['v2'], dt['v1'], test_size=0.2, random_state=42)
# Create the TfidfVectorizer object
vectorizer = TfidfVectorizer(stop_words='english')
```

```python
# Transform the training and testing data
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
# Create the Multinomial Naive Bayes model
model = MultinomialNB()
```

```python
# Train the model
model.fit(X_train, y_train)
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluasi hasilnya dan bandingkan dengan hasil pada Tugas no 2.
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print(classification_report(y_test, y_pred))
```
